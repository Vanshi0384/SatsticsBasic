```python
#Question1 
#Statistics is the science of organising and analysing data and helps in ddecision making process.

#Question 2
#There are two types of statistics:
#1. Descriptive statistics- Consists of organising and summarizing data.
#eg- this can be used when we wnt to summarize large datasets and organise them in a particular given order.
#2. Inferential Statistics- consists of using data to form conclusion. 
#eg- this can be used when conclusions are needed, such as maxima of data needs to be calculated.

#Question 3 
# Different types of data are:

# 1. Quantitative data: Data which is present in numerical form and on which mathematical operations can be performed.
# It is of two types:
# (i) Discrete : It consists of whole no. in a specific range.
# eg- no. of students in a class
# (ii) Continuous: It can consists of whole no. as well as decimal or fraction values.
# eg- height of students in a class
#2. Qualitative Data: Data which consists of categorical data.
# It is of two types:
# (i) Nominal: It consists of categorical data without rank.
# eg- gender, age , Colour
# (ii) Ordinal: It consists of categorical data with ranks.
# eg: Rating of Food(Bad, Okay, Good, delicious)

#Question4:
# (i) Ordinal
#(ii) Nominal
#(iii) Continuous
#(iv) Discrete

#Question 5:
#The levels of measurement in data refer to how precisely a variable is measured and how the values can be interpreted. 
#There are four main levels, each building on the properties of the one before:
# 1.Nominal scale data: This is used for Qualitative or categorical data.
#eg- fav colour, labels etc.
# 2. Ordinal scale data: This is also used for categorical data and here ranking is important but difference can't be measured.
#eg- rating of an app:Good, Satisfactory, bad
#3. Interval scale data: Interval data has equal intervals between each value on the scale.
#This allows you to compare the magnitude of the difference between values.
#eg- Temperature in celcius or farehnit
#4. Ratio scale data:  Ratio data has all the properties of interval data, plus it has a true zero point.
#This means a value of zero truly represents the absence of something being measured.
#eg-  Weight in kilograms/pounds - 0 kg represents no weight

#Question 6:Understanding the levels of measurement of your data is crucial for analyzing it correctly because it determines what kind of information you can glean from it and what statistical methods you can use.
#It's like using the right tool for the job.
#Let's say you analyze customer satisfaction ratings (ordinal data) and find an average rating of 3 stars. This tells you something, but not everything.
#You can't calculate things like an average difference in satisfaction between two groups.
#However, if you had collected purchase data (ratio level - amount spent), you could calculate the average difference in spending between highly satisfied customers (5 stars) and less satisfied customers (1 star).
#This would give you a much richer understanding of customer behavior.
#By understanding the level of your data, you choose the right statistical tests.
#Using the wrong test on the wrong data can lead to misleading or meaningless conclusions.
#So, it's like having the right tools in your toolbox - the level of measurement helps you pick the most appropriate tool for the job of analyzing your data.

#Question 7: 
#Nominal Data:

#Think of labels or categories with no inherent order. It's like sorting socks by color, not size.
#Examples:
#Hair color (blonde, brunette, red)
#Blood type (A, B, AB, O)
#Zip code
#Survey answer choices (Yes/No, Agree/Disagree)

#This data can be ranked or ordered, but the intervals between ranks aren't necessarily equal. Imagine ranking students based on grades (A, B, C), but the gap in knowledge between an A and B student might be different from a B and C student.
#Examples:
#Customer satisfaction ratings (1-5 stars)
#Course grades (A, B, C)
#Military rank (Private, Corporal, Sergeant)

#Question 8:
#Histogram

#Question 9
#1. Descriptive statistics- Consists of organising and summarizing data.
#eg- this can be used when we want to summarize large datasets and organise them in a particular given order.
#2. Inferential Statistics- consists of using data to form conclusion. 
#eg- this can be used when conclusions are needed, such as maxima of data needs to be calculated.

#Question 10:
#Central tendency measures summarize a dataset by representing its "middle" or most typical value.
# 1. Mean: Often referred to as the "average," the mean is calculated by adding all the values in a dataset and dividing by the number of values.
#It's a good all-around measure, but can be sensitive to outliers (extreme values).
#2. Median: The median is the "middle" value when the data is arranged in ascending or descending order. 
#If you have an even number of values, the median is the average of the two middle values. 
#Less sensitive to outliers than the mean.

#3. Mode: The mode is the most frequent value in a dataset. 
#There can be one mode, multiple modes (multimodal data), or even no mode (uniform data).

#Variability: 
#Range: The range is the simplest measure of variability. 
#It's just the difference between the highest and lowest values in the dataset.

#Standard Deviation (SD): The standard deviation (SD) is a more informative measure of variability. 
#It tells you how much, on average, each data point deviates from the mean. A lower SD indicates the data points are closer to the mean, while a higher SD indicates more spread out data.

```
